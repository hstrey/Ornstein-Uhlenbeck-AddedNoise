```python
MCMClist = sorted(glob.glob("MCMCdata/*.csv"))
print(MCMClist)
```

```python
# read first file
MCMCdata = pd.read_csv(MCMClist[0])
for i in range(1,len(MCMClist)):
    moredata = pd.read_csv(MCMClist[i])
    MCMCdata = MCMCdata.append(moredata)
MCMCdata
```

```python
# lets first calculate tau and dtau
MCMCdata['tau'] = -MCMCdata['dt']/np.log(MCMCdata['B'])
MCMCdata['dtau'] = MCMCdata['dt']*MCMCdata['dB']/np.log(MCMCdata['B'])**2/MCMCdata['B']
MCMCdata
```

```python
# now lets average over dt
MCMCmean = MCMCdata.groupby('dt').mean().reset_index()
MCMCmean
```

```python
# lets plot the result
plt.plot(MCMCmean['dt'],MCMCmean['dA'],label="dA")
plt.plot(MCMCmean['dt'],MCMCmean['dtau'],label='dtau')
plt.xlim((0,3))
plt.ylim((0.01,0.2))
plt.legend()
```

```python
# lets plot the result
plt.plot(MCMCmean['dt'],MCMCmean['A'],label="A")
plt.plot(MCMCmean['dt'],MCMCmean['tau'],label='tau')
plt.plot(MCMCmean['dt'],MCMCmean['s'],label='s')
plt.xlim((0,3))
#plt.ylim((0.01,0.2))
plt.legend()
```
