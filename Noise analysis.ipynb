```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


SMALL_SIZE = 16
MEDIUM_SIZE = 18
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

SEED = 35010732 # from random.org
np.random.seed(SEED)

print(plt.style.available)
plt.style.use('seaborn-white')
```

```python
data010 = pd.read_csv("dataN2000/noise010.csv")
data025 = pd.read_csv("dataN2000/noise025.csv")
data050 = pd.read_csv("dataN2000/noise50.csv")
```

```python
data010['tau'] = -data010['delta_t']/np.log(data010['b_mean'])
data025['tau'] = -data025['delta_t']/np.log(data025['b_mean'])
data050['tau'] = -data050['delta_t']/np.log(data050['b_mean'])
```

```python
plt.figure(figsize=(7,5))
plt.plot(data010['delta_t'],data010['stddiff'],label="Pnoise 0.1")
plt.plot(data025['delta_t'],data025['stddiff'],label="Pnoise 0.25")
plt.plot(data050['delta_t'],data050['stddiff'],label="Pnoise 0.5")
plt.axhline(y=np.sqrt(0.1), color='b', linestyle='-.')
plt.axhline(y=np.sqrt(0.25), color='orange', linestyle='-.')
plt.axhline(y=np.sqrt(0.5), color='g', linestyle='-.')
plt.legend()
```

```python
plt.figure(figsize=(7,5))
plt.plot(data010['delta_t'],data010['a_mean'],label="Pnoise 0.1")
plt.plot(data025['delta_t'],data025['a_mean'],label="Pnoise 0.25")
plt.plot(data050['delta_t'],data050['a_mean'],label="Pnoise 0.5")
plt.xlabel(r'$\Delta t/\tau$')
plt.ylabel("A")
plt.legend()
```

```python
plt.figure(figsize=(7,5))
plt.plot(data010['delta_t'],data010['tau'],label="Pnoise 0.1")
plt.plot(data025['delta_t'],data025['tau'],label="Pnoise 0.25")
plt.plot(data050['delta_t'],data050['tau'],label="Pnoise 0.5")
plt.xlabel(r'$\Delta t/\tau$')
plt.ylabel(r'$\tau$')
plt.legend()
```
